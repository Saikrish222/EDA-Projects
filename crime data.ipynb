```python
under_20_data = data[data["Vict Age"] < 20]

# Histogram
plt.subplot(1, 2, 1)
sns.histplot(under_20_data["Vict Age"], bins=10, kde=False, color='skyblue')
plt.title('Distribution of Victim Ages Under 20')
plt.xlabel('Age')
plt.ylabel('Frequency')

# Show the plots
plt.tight_layout()
plt.show()
```

```python
#lineplot
sns.lineplot(x="Vict Age", y="Premis Desc", data=data)
```

```python
#histplot
sns.histplot(data['Vict Age'].dropna(), bins=30, kde=True)
plt.title('Distribution of Victim Ages')
plt.xlabel('Frequency')
plt.ylabel('Age')
plt.show()
```

```python
#pieplot
data['Vict Sex'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
```

```python
#heatmap
numeric_data = data.select_dtypes(include=['float64', 'int64'])
sns.heatmap(numeric_data.corr(),  cmap='coolwarm',annot=True)
plt.title('Correlation ')
plt.show()
```

```python
#scatterplot
sns.scatterplot(x='Vict Age', y='Vict Sex', data=data)
plt.title('Scatterplot of Vict Age vs.Vict Sex')
plt.xlabel('Vict Age')
plt.ylabel('Vict Sex')
plt.show()
```

```python
#contplot
sns.countplot(data=data, x='Status Desc', color= 'Red')
plt.title('Distribution of Crime Status')
plt.xlabel('Status Description')
plt.ylabel('Number of Crimes')
plt.show()
```
