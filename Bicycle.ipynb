```python
#lineplot
import seaborn as sns 
sns.lineplot(x="weekday", y="workingday", data=data)
```

```python
#boxplot
sns.boxplot (x = "weekday", y = "weathersit", data=data)
```

```python
#violinplot
sns.violinplot(x = 'casual', y='cnt', data=data)
```

```python
#pie plot
data['weekday'].value_counts().plot.pie(autopct='%1.1f%%',startangle=90)
plt.title("weekday")
```

```python
#histplots
plt.figure(figsize=(5, 4))
sns.histplot(data['weekday'], bins=30, kde=True)
plt.title('Distribution of weekday ')
plt.show()
```
