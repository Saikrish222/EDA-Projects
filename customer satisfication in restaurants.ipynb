```python
#countplot
plt.figure(figsize=(8, 4))
sns.countplot(x=data['HighSatisfaction'])
plt.title('Distribution of HighSatisfaction')
plt.show()
```

```python
#histplot
numeric_columns = ['Age', 'Income', 'AverageSpend', 'WaitTime', 'ServiceRating', 'FoodRating', 'AmbianceRating']
for column in numeric_columns:
    plt.figure(figsize=(8, 4))
    sns.histplot(data[column].dropna(), kde=True)
    plt.title(f'Distribution of {column}')
    plt.show()
```

```python
#heatmap
data_encoded = pd.get_dummies(data, drop_first=True)

plt.figure(figsize=(12, 8))
sns.heatmap(data_encoded.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
#heatmap
corr_with_target = data_encoded.corr()['HighSatisfaction'].sort_values(ascending=False)
plt.figure(figsize=(10, 6))
sns.barplot(x=corr_with_target.index, y=corr_with_target.values)
plt.xticks(rotation=90)
plt.title('Correlation of Features with HighSatisfaction')
plt.show()
```

```python
data['Gender'].value_counts().plot.pie(autopct='%1.1f%%',startangle=180)
plt.title("Gender")
```

```python
sns.scatterplot(x='Age', y='GroupSize', data=data)
plt.title('Scatterplot of Age vs.GroupSize')
plt.xlabel('Age')
plt.ylabel('GroupSize')
plt.show()
```
