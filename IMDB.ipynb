```python
#lineplot
sns.lineplot(x='Year', y='IMDb Rating', data=data)
plt.title('Lineplot of Year vs. IMDb Rating')
plt.show()
```

```python
#barplot
sns.barplot(x='Genre', y='IMDb Rating', data=data)
plt.title('Average IMDb Rating by Genre')
plt.show()
```

```python
#hist
sns.histplot(data['IMDb Rating'])
plt.title('Distribution of IMDb Ratings')
plt.show()
```

```python
#scatterplot
sns.scatterplot(x='Duration (minutes)', y='IMDb Rating', data=data)
plt.title('Duration vs. IMDb Rating')
plt.show()
```

```python
#boxplot
sns.boxplot(x='IMDb Rating', y='Genre', data=data)
plt.title('IMDb Rating by Genre')
plt.show()
```

```python
#violinplot
sns.violinplot(x='Genre', y='IMDb Rating', data=data)
plt.title('IMDb Rating by Genre')
plt.show()
```
