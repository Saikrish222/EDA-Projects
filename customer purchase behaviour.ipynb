```python
# anaylyze the data
data = pd.read_csv(r"C:\Users\harun\Downloads\customer_purchase_data.csv")
data.head()
```

```python
#lineplot
import seaborn as sns 
sns.lineplot(x="ProductCategory", y="NumberOfPurchases", data=data)
```

```python
#barplot
sns.barplot(x = 'ProductCategory',y = 'NumberOfPurchases',data=data)
```

```python
#boxplot
sns.boxplot (x = "PurchaseStatus", y = "NumberOfPurchases", data=data)
```

```python
#violinplot
sns.violinplot(x = 'DiscountsAvailed', y='PurchaseStatus', data=data)
```

```python
#heatmap
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
plt.show
```

```python
#pie plot
data['PurchaseStatus'].value_counts().plot.pie(autopct='%1.1f%%',startangle=180)
plt.title("PurchaseStatus")
```

```python
#histplots
plt.figure(figsize=(5, 4))
sns.histplot(data['NumberOfPurchases'], bins=30, kde=True)
plt.title('Distribution of NumberOfPurchases')
plt.show()
```

```python
# Creating the relplot
sns.relplot(
    x='AnnualIncome', 
    y='TimeSpentOnWebsite',
    hue='PurchaseStatus', 
    style='Gender',
    size='LoyaltyProgram',
    sizes=(5, 20),
    col='ProductCategory',
    data=data
)

plt.suptitle('Relationship Between Annual Income and Time Spent On Website by Purchase Status and Gender', y=1.02)
plt.show()
```
