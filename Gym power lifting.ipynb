```python
data.nlargest(n = 5, columns = ['Amount Lifted (kg)'])
#barplot
plt.figure(figsize=(10, 6))
sns.barplot(x='Lifter Name', y='Amount Lifted (kg)', data=data, palette='viridis',hue='Lifter Name')
plt.xlabel('Lifter Name')
plt.ylabel('Amount Lifted (kg)')
plt.title('Top 5 Maximum Amounts Lifted')
plt.show()
```

```python
df = data[data['Lift Type'] == 'Deadlift']
print(df)
```

```python
df =data['Lift Type'].value_counts()
#pieplot
plt.figure(figsize=(6, 4))
plt.pie(df,labels=df.index, autopct='%1.1f%%', colors=sns.color_palette('muted'), startangle=90)
plt.title('Distribution of Lift Type')
plt.show()
```

```python
#boxplot
sns.boxplot(x='Lift Type', y='Weight Class', data=data, palette='Set2',hue='Lift Type')
plt.xlabel('Lift Type')
plt.ylabel('Weight Class')
plt.title('Comparison of Weight Lifted by Lift Type')
plt.show()
```

```python
#lineplot
sns.lineplot(x='Age', y='Amount Lifted (kg)', data=data, marker='o', color='r')
plt.title('Amount Lifted by Age')
plt.xlabel('Age')
plt.ylabel('Amount Lifted (kg)')
plt.grid(True)
plt.show()
```
